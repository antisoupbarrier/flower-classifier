```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.applications.densenet import preprocess_input
import matplotlib.pyplot as plt
```

```python
# Taken from classes within training script (Note: Class order must match training class order)

classes=['california_poppy',
         'iris',
         'magnolia',
         'bellflower',
         'astilbe',
         'water_lily',
         'dandelion',
         'sunflower',
         'calendula',
         'common_daisy',
         'tulip',
         'black_eyed_susan',
         'coreopsis',
         'carnation',
         'daffodil',
         'rose'
        ]
```

```python
flowers_path='./flowers'
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    flowers_path,
    validation_split=0.2,
    subset="validation",         
    seed=42,
    image_size=(150, 150),
    batch_size=64,
    label_mode="categorical",
    class_names=classes
)
```

```python
def preprocess_image(image, label):
    image = preprocess_input(image)
    return image, label
```

```python
model = tf.keras.models.load_model('densenet201_v3_15_0.913.h5')
```

```python
val_ds = val_ds.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
```

```python
results = model.evaluate(val_ds)
```

```python
# Plots images from dataset with real and predicted classes 

plt.figure(figsize=(20, 10))
for images, labels in val_ds.take(1):
    for i in range(40):
        ax = plt.subplot(8, 5, i + 1)
        
        img_array = images[i].numpy().astype("float")
        
        prediction = model.predict(np.array([img_array]))
        prediction_name = classes[np.argmax(prediction)]
        real_name = classes[np.argmax(labels[i])]
        
        # Reverse the [-1,1] scaling for display
        img_array = (img_array + 1) / 2.0
        plt.imshow(img_array)
        
        if prediction_name == real_name:
            plt.title(f'real: {real_name}\npred:{prediction_name}', fontdict={'color': 'g'})
        else:
            plt.title(f'real: {real_name}\npred:{prediction_name}', fontdict={'color': 'r'})
        
        plt.axis("off")

plt.show()
```
